```python
import pandas as pd

df = pd.read_csv("features.csv")
df.head()
```

```python
df.info() #datatypes and non null counts
```

```python
df.describe() #summary of statistics
```

```python
#deleting column jitter_local and shimmer_local
df = df.drop(columns = ['jitter_local','shimmer_local'])
```

```python
df.isnull().sum() #checking missing values
```

```python
# checking for duplicated columns
df.duplicated().sum()
```

```python
from sklearn.preprocessing import LabelEncoder

if "recording_type" in df.columns:
    le = LabelEncoder()
    df["recording_type"] = le.fit_transform(df["recording_type"])

```

```python
from sklearn.preprocessing import StandardScaler

X = df.drop(columns=["label", "filename"], errors="ignore")
y = df["label"]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
#Histogram of features
numeric_df = df.select_dtypes(include=['float64', 'int64'])
numeric_df.hist(figsize=(15,12));

```

```python
# Correlation Heatmap
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12,8))
sns.heatmap(df.select_dtypes(include=['float64', 'int64']).corr(), cmap="coolwarm", annot=False)

```

```python
# Class distribution
df['label'].value_counts().plot(kind='bar');
```
